```python
f['language'].value_counts()
```

```python
x=['English', 'Spanish', 'French', 'German', 'Slavic','Other-Indo-European',
   'Chinese', 'Arabic','Japanese/Turkish/Finnish/Magyar', 'Others']
```

```python
plt.figure(figsize=(10,7))
plt.pie(f['language'].value_counts(),autopct='%1.2f%%', pctdistance=1.2);
plt.legend(bbox_to_anchor=(1, 1),labels=x);
```

### From above dig. we can conclude that english is the most preferred language in all countries.

```python
plt.figure(figsize=(12,5))
sns.countplot(x=f['topleft'],order=f['topleft'].value_counts().index.sort_values())
plt.xticks(rotation='vertical')
plt.xlabel('color',fontsize=12)
plt.ylabel('No.of countries',fontsize=12);
plt.title('Topleft color vs No.of Countries ',fontsize=16);
```

```python
x=f[f['area']>=1000][['country_name','area']]
x.sort_values('area')
```

```python
f[f['area']>8000]
```

```python
#area	in thousands of square km
plt.figure(figsize=(10,5));
sns.barplot(x=f[f['area']>=1000]['country_name'],y=f['area'],palette='rainbow');
plt.xticks(rotation='vertical');
plt.xlabel('country_name',fontsize=12);
plt.ylabel('area',fontsize=12);
plt.title('country_name vs area',fontsize=14);
```

### From above Graph we conclude that Canada has the highest area.

```python
sns.set(rc = {'figure.figsize':(10,10)})
sns.barplot(x=f['population'],y=f[f['population']>20]['country_name']);
```

### From above graph we can conclude that China have a max. population
